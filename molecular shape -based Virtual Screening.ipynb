```python
Molecular shape-based Virtual Screening
```

```python
#!/usr/bin/python2.7

import oddt
from oddt.shape import usr, usr_similarity
```

```python
#load query file
query = next(oddt.toolkit.readfile('mol2', 'query.mol2'))
```

```python
query
```

```python
#load molecular database
database = list(oddt.toolkit.readfile('mol2', 'molecules.mol2'))
```

```python
#find similar molecules
results = []
query_shape = usr(query)
for mol in database:
	mol_shape = usr(mol)
	similarity = usr_similarity(query_shape, mol_shape)	
	if similarity > 0.8:
		results.append(mol)
```

```python
results
```

```python
#output single mol2 file
for i in range(len(results)):
	results[i].write("mol2",str(i)+".mol2")
```
